```python
#importing libraries required
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/mars-orbiter/RF-lab/main/RF1b/rf1b.csv'
dataset = pd.read_csv(url)
print(dataset)
```

```python
# accessing the different columns of the csv dataset I made above
dataset.columns = ["Vi","Vr2","Vr4","Vr8"]
#storing values for each column into a separate arrays
V_I = np.array(dataset.Vi)                     # Values of V_I
V_R_2 = np.array(dataset.Vr2)                  # Values of V_R for VSWR = 2
V_R_4 = np.array(dataset.Vr4)                  # Values of V_R for VSWR = 4
V_R_8 = np.array(dataset.Vr8)                  # Values of V_R for VSWR = 8
```

```python
#plotting the graph between V_I and V_R
# For VSWR = 2
plt.plot(V_I, V_R_2,'b.-')
plt.scatter(V_I, V_R_2, s = 10)

#plotting
plt.xticks(np.arange(0,51,5))
plt.yticks(np.arange(0,20,3))
plt.title("Graph between $V_I$ and $V_R$")
plt.xlabel(r'$V_I$ (in V)$\longrightarrow$')                #x label
plt.ylabel(r'$V_R$ (in V)$\longrightarrow$')                #y label
plt.show()
```

```python
#plotting the graph between V_I and V_R
# For VSWR = 4
plt.plot(V_I, V_R_4,'r.-')
plt.scatter(V_I, V_R_4, s = 10)

#plotting
plt.xticks(np.arange(0,51,5))
plt.yticks(np.arange(0,31,3))
plt.title("Graph between $V_I$ and $V_R$")
plt.xlabel(r'$V_I$ (in V)$\longrightarrow$')                #x label
plt.ylabel(r'$V_R$ (in V)$\longrightarrow$')                #y label
plt.show()
```

```python
#plotting the graph between V_I and V_R
# For VSWR = 8
plt.plot(V_I, V_R_8,'g.-')
plt.scatter(V_I, V_R_8, s = 10)

#plotting
plt.xticks(np.arange(0,51,5))
plt.yticks(np.arange(0,41,3))
plt.title("Graph between $V_I$ and $V_R$")
plt.xlabel(r'$V_I$ (in V)$\longrightarrow$')                #x label
plt.ylabel(r'$V_R$ (in V)$\longrightarrow$')                #y label
plt.show()
```

```python
#plotting the graph between V_I and V_R
# For VSWR = 2
plt.plot(V_I, V_R_2,'b.-',label = "VSWR = 2")
plt.scatter(V_I, V_R_2,s = 10)

# For VSWR = 4
plt.plot(V_I, V_R_4,'r.-',label = "VSWR = 4")
plt.scatter(V_I, V_R_4, s = 10)

# For VSWR = 8
plt.plot(V_I, V_R_8,'g.-',label = "VSWR = 8")
plt.scatter(V_I, V_R_8, s = 10)

#plotting
plt.xticks(np.arange(0,51,5))
plt.yticks(np.arange(0,41,3))
plt.title("Graph between $V_I$ and $V_R$")
plt.xlabel(r'$V_I$ (in V)$\longrightarrow$')                #x label
plt.ylabel(r'$V_R$ (in V)$\longrightarrow$')                #y label
plt.legend()
plt.show()
```
