```python
#importing libraries required
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# LOW PASS FILTER
LPF = 'https://raw.githubusercontent.com/mars-orbiter/RF-lab/main/lpf.csv'
LPFDATA = pd.read_csv(LPF, encoding='cp1252')
print(LPFDATA)

# accessing the different columns of the csv dataset I made above
LPFDATA.columns = ["Serial_No.",  "Frequency_Hz",  "Magnitude_dB","Phase_theta" , "V_out"]
#storing values for each column into a separate list
Phase = np.array(LPFDATA.Phase_theta)
Frequency = np.array(LPFDATA.Frequency_Hz)
Magnitude = np.array(LPFDATA.Magnitude_dB)

plt.subplots(figsize=(15, 5))
plt.suptitle('Graphs for Low Pass Filter', fontsize=20)
# GRAPH BETWEEN FREQUENCY AND PHASE (FREQUENCY IN LOGSCALE)
plt.subplot(1, 2, 1)
plt.scatter(Frequency,Phase,s=10)
plt.plot(Frequency,Phase)
plt.xscale('log')
plt.xlabel('Frequency (Hz) (logscale)$\longrightarrow$')
plt.ylabel('Phase(theta) $\longrightarrow$')
plt.title('Frequency vs Phase', fontsize = 15)

# GRAPH BETWEEN FREQUENCY AND Magnitude (FREQUENCY IN LOGSCALE)
plt.subplot(1, 2, 2)
plt.scatter(Frequency,Magnitude,s=10)
plt.plot(Frequency,Magnitude)
plt.xscale('log')
plt.xlabel('Frequency (Hz) (logscale)$\longrightarrow$')
plt.ylabel('Magnitude(dB) $\longrightarrow$')
plt.title('Frequency vs Magnitude', fontsize = 15)
plt.show()
```

```python
# HIGH PASS FILTER
HPF = 'https://raw.githubusercontent.com/mars-orbiter/RF-lab/main/hpf.csv'
HPFDATA = pd.read_csv(HPF, encoding='cp1252')
print(HPFDATA)

# accessing the different columns of the csv dataset I made above
HPFDATA.columns = ["Serial_No.",  "Frequency_Hz",  "Magnitude_dB","Phase_theta" , "V_out"]
#storing values for each column into a separate list
Phase1 = np.array(HPFDATA.Phase_theta)
Frequency1 = np.array(HPFDATA.Frequency_Hz)
Magnitude1 = np.array(HPFDATA.Magnitude_dB)


plt.subplots(figsize=(15, 5))
plt.suptitle('Graphs for High Pass Filter', fontsize=20)
# GRAPH BETWEEN FREQUENCY AND PHASE (FREQUENCY IN LOGSCALE)
plt.subplot(1, 2, 1)
plt.scatter(Frequency1,Phase1,s=10,color ='orange')
plt.plot(Frequency1,Phase1,color ='orange')
plt.xscale('log')
plt.xlabel('Frequency (Hz) (logscale)$\longrightarrow$')
plt.ylabel('Phase(theta) $\longrightarrow$')
plt.title('Frequency vs Phase', fontsize = 15)

# GRAPH BETWEEN FREQUENCY AND Magnitude (FREQUENCY IN LOGSCALE)
plt.subplot(1, 2, 2)
plt.scatter(Frequency1,Magnitude1,s=10,color ='orange')
plt.plot(Frequency1,Magnitude1,color ='orange')
plt.xscale('log')
plt.xlabel('Frequency (Hz) (logscale)$\longrightarrow$')
plt.ylabel('Magnitude(dB) $\longrightarrow$')
plt.title('Frequency vs Magnitude', fontsize = 15)
plt.show()
```
